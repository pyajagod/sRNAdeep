```python
import pandas as pd
import csv
```

### 1. convert kmer sequence to original sequences（for do blastn）

```python
def kmer_to_sequence(kmer_list, k=3):
    """
     Restore the k-mer list generated by the sliding window to the original sequence
    :param kmer_list: the input k-mer list
    :param k: length of each k-mer, default is 3
    :return: the original sequence to be restored
    """
    if not kmer_list:
        return ""
    
    # Initialisation sequence for the first k-mer
    original_sequence = kmer_list[0]
    
    # Starting with the second k-mer, append the last character of each k-mer to the original sequence
    for kmer in kmer_list[1:]:
        original_sequence += kmer[-1]  # Add the last character of each k-mer
    
    return original_sequence
```

```python
def csv_to_fasta(input_csv, output_fasta, tag='train_', k = 3):
    with open(input_csv, 'r') as csv_file, open(output_fasta, 'w') as fasta_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            kmer_sequence = row['sequence']
            if k > 0:
                original_seq = kmer_to_sequence(kmer_sequence.split(' '))
            else:
                original_seq = kmer_sequence
            fasta_file.write(f'>{tag}RNA{csv_reader.line_num-1}\n{original_seq}\n')
```

```python
input_csv_file = './data/test.csv'
output_fasta_file = ''

csv_to_fasta(input_csv_file, output_fasta_file, 'test_', k = 0)
```
